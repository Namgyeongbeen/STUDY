```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# pandas-visualization의 df1, df2, df3 데이터 사용.

df1 = pd.read_csv('df1.csv')
df2 = pd.read_csv('df2.csv')
df3 = pd.read_csv('df3.csv')
```

```python
display(df1.head())
display(df2.head())
display(df3.head())
```

# 기본 histogram 시각화

```python
# 기본적 히스토그램 시각화.
df1.A.plot.hist()
```

```python
# 각 막대에 테두리 색 넣어주기
df1.A.plot.hist(edgecolor='black')
plt.show()
```

```python
# 히스토그램 막대를 축의 가장자리에 맞추기.
# autosclae. 파라미터 중 both => x,y축 모두의 가장자리에 맞춤
df1.A.plot.hist(edgecolor='black').autoscale(enable=True, axis='both', tight=True)
plt.show()
```

```python
# 히스토그램 막대 수 늘리기
df1.A.plot.hist(bins=20, edgecolor='black')
plt.show()
```

```python
# 격자무늬 추가하기
# 파라미터 grid=True 추가
df1.A.plot.hist(bins=20, edgecolor='black', grid=True)
plt.show()
```

## plot.area

```python
# 영역으로 채우기.
# 기본적으로 stack 되어 있는 상태임.
df2.plot.area()
plt.show()
```

```python
# 투명도 추가하기.
# alpha 파라미터 추가. 0.0~1.0
df2.plot.area(alpha=0.5)
plt.show()
```

```python
# stack 된 값이 아니라 각각의 영역을 겹쳐서라도 각각의 값을 보고 싶은 경우엔
# stacked=False 파라미터 추가
df2.plot.area(alpha=0.5, stacked=False)
plt.show()
```

## plot.scatter

```python
# 변수간의 값을 비교하는데 유용하게 쓰임.
df1.plot.scatter(x='A', y='B')
plt.show()
```

```python
# 컬럼 C에 대한 컬럼 A와 컬럼 B의 관계를 나타내고 싶을 때. 특히, 컬럼 C에 색을 넣고 싶은 경우?
# c 파라미터 추가. color의 약자.
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
plt.show()

# A,B,C 사이에 관계가 있음을 나타낸다.
```

```python
# 크기를 다양하게 할 수도 있다.
# s 파라미터 추가. size의 약자.
df1.plot.scatter(x='A', y='B', s='C')
plt.show()

# /usr/local/lib/python3.10/dist-packages/matplotlib/collections.py:963: RuntimeWarning: invalid value encountered in sqrt scale = np.sqrt(self._sizes) * dpi / 72.0 * self._factor
# 위와 같은 메시지는, size를 조절한 C열의 값 중 음수나 무한대가 있거나, 'C' 자체를 문자로 인식했기 때문에 출력된다. 그러면 크기를 나타낼 수 없으니까.
# 이런 경우엔 그냥 한번 더 실행하거나, df1['C']라는 식으로 df1의 C 컬럼의 값을 사용하라고 명확하게 지정해주면 된다.
```

```python
# 크기 비율을 늘려보자.
# 늘리고 싶은 만큼 정수를 곱해주면 된다.
df1.plot.scatter(x='A', y='B', s=df1['C']*50)
plt.show()
```

```python
# 투명도를 추가해서 얼마나 겹쳐있는지 보고 싶다면?
# area plot에서처럼 alpha 추가
df1.plot.scatter(x='A', y='B', s=df1['C']*50, alpha=0.3)
plt.show()

# 물론 색 변경도 가능하다.
```

## plot.box

```python
df2.plot.box()
plt.show()

print('='*100)
df2.boxplot()
plt.show()

# boxplot은 각 컬럼 값의 분포를 나타냄.
# 박스 위쪽 부터 위쪽 바깥의 선까지는 상위 25%가 분포해있고, 반대로 아래쪽부터 아래쪽 바깥의 선까지는 하위 25%가 분포해있음
# 박스 안에는 그 외 값(중간 50%)이 분포함.
# 박스 안의 초록 선은 중앙값. 즉 초록 선 위쪽은 상위 50%, 아래쪽은 하위 50%
```

## 사용자 정의

```python
df2 = pd.read_csv('df2.csv')

# 전체 그래프 크기, 선 모양(ls), 선 색깔(c), 선 굵기(lw) 튜닝
df2.c.plot.line(figsize=(15, 3), ls='-.', c='red', lw=3)
```

```python
# title, x label, y label 지정
title = 'my plot title'
xlabel = 'my x data'
ylabel = 'my y data'

ax = df2.c.plot.line(figsize=(15, 3), ls='-.', c='red', lw=3)
ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
plt.show()
```

```python
# 범례 지정
ax = df2.plot()
ax.legend(loc='best')
plt.show()

# 범례를 밖으로 옮기려면?
ax = df2.plot()
ax.legend(loc='best', bbox_to_anchor=(1.0, 1.0))
plt.show()
# 이 경우엔 legend가 처음 나타나는 위치인 'best'에 영향을 받는다. loc이 어디냐에 따라 다름.
```
